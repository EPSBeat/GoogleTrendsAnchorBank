# `gtab` example with categories

Below we build a gtab with categories! List of all available categories is available [here](https://github.com/pat310/google-trends-api/wiki/Google-Trends-Categories).

The original list which we use to build anchorbanks may not be appropriate for querying other categoires, please use the alternate candidate file provided `550_cities_and_countries.txt.` Note that below, we also made the number of anchors smaller, this can be helpful to speed things up.

````python
import gtab

t = gtab.GTAB()
t.set_options(pytrends_config={"geo": "DE", "timeframe": "2018-03-05 2021-05-05", 'cat': "71"},
              gtab_config= {"anchor_candidates_file": "550_cities_and_countries.txt", "num_anchors": 100})
````

````python
t.create_anchorbank(verbose=True) # takes a while to run since it queries Google Trends.
````

````python
t.set_active_gtab("google_anchorbank_geo=DE_timeframe=2018-03-05 2021-05-05_cat=71.tsv")
````

````python
example_query = t.new_query("Christmas")
````

````python
import matplotlib.pyplot as plt
plt.figure(figsize=(14, 4))
plt.plot(example_query.max_ratio)
plt.title("Queries for Christmas in the 'Food' category (cat=71)")
plt.show()
````

When using categories, always ensure that you are setting the options correctly, specifying that you want that category (when you make the anchorbank, but also when you query it).
